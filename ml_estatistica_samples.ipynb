```python
import pandas as pd

houses = pd.read_csv('casas_1000.csv')
houses
```

```python
houses.info()
```

```python
houses.describe()
```

```python
import numpy as np

houses.loc[houses.sample(frac=0.1).index, 'número_de_quartos'] = np.nan

houses.isnull().sum()
```

```python
mediana_numero_quartos = houses['número_de_quartos'].median()
houses['número_de_quartos'].fillna(mediana_numero_quartos, inplace=True)
houses.isnull().sum()
```

```python
houses['número_de_quartos'].describe()
```

```python

Q1 = houses['tamanho_da_casa'].quantile(0.25)
Q3 = houses['tamanho_da_casa'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

houses_filtered = houses[(houses['tamanho_da_casa'] >= lower_bound) & (houses['tamanho_da_casa'] <= upper_bound)]
```
