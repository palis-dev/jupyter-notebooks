```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar um array de tempo de 0 a 5 segundos
t = np.linspace(0, 5, 200)

# Funo de posio: s(t) = t
s = t**2

# Derivada da posio (velocidade): v(t) = 2t
v = 2 * t

# Criar o grfico
plt.figure(figsize=(10,6))
plt.plot(t, s, label='Posio s(t) = t', color='blue')
plt.plot(t, v, label='Velocidade v(t) = 2t', color='red', linestyle='--')
plt.title('Curva do Movimento e sua Derivada (Velocidade)')
plt.xlabel('Tempo (t)')
plt.ylabel('Espao')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gera um conjunto de valores de probabilidade predita entre 0.01 e 0.99
x = np.linspace(0.01, 0.99, 200)

# Para o caso y = 1, a funo de perda de cross entropy :
loss = -np.log(x)

# Clculo da derivada da loss em funo de x (predicted probability)
dloss_dx = -1 / x

# Escolhe um ponto especfico para calcular a reta tangente
x0 = 0.5
L0 = -np.log(x0)  # valor da loss em x0
slope = -1 / x0  # derivada em x0

# Calcula os valores da reta tangente
tangent_line = L0 + slope * (x - x0)

# Cria o grfico
plt.figure(figsize=(10, 6))
plt.plot(x, loss, label='Cross Entropy Loss (y=1)', color='blue')
plt.plot(x, tangent_line, label=f'Tangente em  = {x0}', color='red', linestyle='--')
plt.scatter([x0], [L0], color='black', zorder=5)  # Marca o ponto de tangncia
plt.xlabel('Probabilidade Predita ()')
plt.ylabel('Loss')
plt.title('Clculo da Derivada na Cross Entropy Loss')
plt.legend()
plt.grid(True)
plt.show()
```
