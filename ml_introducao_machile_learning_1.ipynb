```python
from fastai.tabular.all import *
from numpy.random import normal, seed, uniform
from ipywidgets import interact, FloatSlider
from functools import partial
```

```python
np.random.seed(42)
def noise(x, scale): return normal(scale=scale, size=x.shape)
def add_noise(x, mult, add): return x * (1+noise(x,mult) + noise(x, add))
def quad(a,b,c,x): return a*x**2 + b*x + c
def mk_quad(a,b,c): return partial(quad, a,b,c)
def quad_mse(params):
  f = mk_quad(*params)
  return mse(f(x), y)

def plot_function(func, xlim=(-10, 10), ylim=None, num_points=100, title="Function Plot"):
    """
    Plots a given mathematical function.

    Parameters:
        func (function): The mathematical function to plot.
        xlim (tuple): The range of x values (default is (-10, 10)).
        ylim (tuple): The range of y values (default is None, which auto-scales).
        num_points (int): Number of points to calculate for the plot (default is 100).
    """
    # Generate x values
    x = np.linspace(xlim[0], xlim[1], num_points)
    # Calculate y values
    y = func(x)

    # Create the plot
    plt.plot(x, y, label="Function")
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    if ylim:
        plt.ylim(ylim)
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()
```
